```python
import pandas as pd
```

```python
df=pd.read_csv("covid_19_india_dataset.csv")
```

```python
#plotting out of total cases how many have recovered and how many are remaining.
#Overall 
sum_of_total_cases=len(df)
print(sum_of_total_cases)

recovered_cases=len(df.loc[df['Current Status']=='Recovered'])
print(recovered_cases)

hospitalized_cases=len(df.loc[df['Current Status']=='Hospitalized'])
print(hospitalized_cases)

deceased_cases=len(df.loc[df['Current Status']=='Deceased'])
print(deceased_cases)

migrated_cases=len(df.loc[df['Current Status']=='Migrated'])
print(migrated_cases)

import matplotlib.pyplot as plt
import numpy 

# Data to plot
labels =['Recovered_cases','Hospitalized_cases','Deceased_cases','Migrated_cases']
sizes = [181,2000,100,20]
colors = ['lightyellow','green', 'red', 'blue']
#plot
plt.pie(sizes,labels=labels,colors=colors,startangle=0)

#plt.legend(labels,loc='upper center', bbox_to_anchor=(1.05, 0.8), shadow=True, ncol=1)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#how the cases have evolved over time 

# Initialize a Figure 
fig= plt.figure(figsize=(20,8))

# Set up Axes
ax = fig.add_subplot(111)     #mean 1,1,1

x=df['Date Announced']
y=df['Patient Number']

ax.bar(x,y)

# rotate and align the tick labels so they look better
fig.autofmt_xdate()

#Add titile and label for axes
ax.set(title="Rise of COVID'19 over time", xlabel="Date Announced", ylabel="Patient count")


```
